# Mapper examples

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from sklearn.datasets import make_circles, make_moons

import persistence
```

```python
X, y = make_circles(factor=0.3, random_state=123)
X.shape, y.shape
```

```python
plt.title("Ground truth labels")
for label in np.unique(y):
    plt.scatter(*X[np.where(y == label)[0]].T)
```

```python
clusters = persistence.dbscan(X, epsilon=0.2, min_points=2)
clusters
```

```python
plt.title("DBSCAN cluster labels")
for label in np.unique(clusters):
    plt.scatter(*X[np.where(clusters == label)[0]].T)
```

```python
graph = persistence.mapper(X, n_divisions=5, epsilon=0.2, min_points=2)
```

```python
G = nx.Graph()
for node, neighbors in graph.items():
    for neighbor in neighbors:
        G.add_edge(node, neighbor)
```

```python
nx.draw(G)
```
